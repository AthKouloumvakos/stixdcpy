```python
import stix
import numpy as np
from matplotlib import pyplot as plt
%matplotlib notebook
```

```python
res= stix.search('2020-06-21T18:00:00','2020-06-21T23:30:00','lc')
#res.info()
```

```python
data=res[1].data
```

```python
lc_data=data['COUNTS'].transpose()

t=[4*x for x in range(len(lc_data[0]))]
```

```python
for i in range(5):
    plt.plot(t, lc_data[i])
plt.xlabel('Time (s)')
plt.ylabel('Counts')
plt.title('STIX light curves')
```
