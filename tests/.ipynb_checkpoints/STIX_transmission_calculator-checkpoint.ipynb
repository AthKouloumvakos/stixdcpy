```python
#attenuator=False
def get_trans(detector_id):
    energies=np.linspace(3,150,10000)
    ebins=energies.reshape((-1,2))
    plt.close('all')
    stran=tr.Transmission(material_version=1)
    
    
    factors_new_mean_wo_att=stran.get_detector_transmission(detector_id,
                                  ebins, attenuator=False)
    
  
    factors_new_mean_att=stran.get_detector_transmission(detector_id,
                                  ebins, attenuator=True)
    print("\n\n##Switch to old materials")
    stran2=tr.Transmission(material_version=0)


    
    factors_old_wo_att=stran2.get_detector_transmission(detector_id,
                                  ebins, attenuator=False)
    factors_old_att=stran2.get_detector_transmission(detector_id,
                                  ebins, attenuator=True)
    
    
    stran3=tr.Transmission(material_version=2)


    
    #factors_new_max_wo_att=stran3.get_detector_transmission(detector_id,
    #                              ebins, attenuator=False)
    factors_new_max_att=stran3.get_detector_transmission(detector_id,
                                  ebins, attenuator=True)
    

    plt.figure()
    ex=np.mean(ebins,axis=1)
    plt.plot(ex, factors_old_wo_att*100, label=f'Old')
    plt.plot(ex, factors_new_mean_wo_att*100, label=f'New')
    plt.xlabel('Energy (keV)')
    plt.ylabel('Transmission (%)')
    plt.xscale('log')
    plt.legend(loc='upper left')
    plt.title(f'Transmission coefficient of Collimator #{detector_id} - without attenuator')
    #plt.yscale('log')
    plt.grid('on')
    
    
    plt.figure()
    plt.plot(ex, factors_old_att*100, label=f'Old')
    
    plt.plot(ex, factors_new_mean_att*100, label=f'New materials (mean mass fraction)')
    
    
    plt.plot(ex, factors_new_max_att*100, label=f'New materials (max mass fraction)')
    
    plt.xlabel('Energy (keV)')
    plt.ylabel('Transmission (%)')
    plt.xscale('log')
    plt.legend()#loc='upper left')
    plt.title(f'Transmission coefficient of Collimator #{detector_id} with attenuator ')
    plt.yscale('log')
    plt.grid('on')
    





#interact_manual(get_trans, detector_id=widgets.IntSlider(min=0,max=32,value=26,step=1)    )
get_trans(26)
        
```
